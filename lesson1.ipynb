```python
import torch
```

```python
x1 = torch.Tensor([1.0])
x2 = torch.Tensor([0.0])
w1 = torch.Tensor([-3.0])
w2 = torch.Tensor([2.0])
b = torch.Tensor([6.8813735])

x1.requires_grad = True
x2.requires_grad = True
w1.requires_grad = True
w2.requires_grad = True
b.requires_grad = True

n = w1*x1 + w2*x2 + b
o = torch.tanh(n)

o.backward()

print(o)
print(o.data)
print(o.data.item())
print(o.item())
```

```python
print(x1.grad.item())
print(x2.grad.item())
print(w1.grad.item())
print(w2.grad.item())

n.retain_grad()
o.retain_grad()
print(n.grad)
print(o.grad)
```
